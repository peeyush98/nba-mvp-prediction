```python
# importing required libraries
import numpy as np
import pandas as pd
```

# Cleaning MVPS Data

```python
# Reading the mvps data
mvps = pd.read_csv("mvps.csv")
```

```python
mvps.head()
```

```python
# Selecting only necessary columns
mvps = mvps[["Player", "Year", "Pts Won", "Pts Max", "Share"]]
```

Got rid of the rest of the columns because those are already present in the players data.

```python
mvps.head()
```

# Cleaning Players Data

```python
# Reading the players data
players = pd.read_csv("players.csv")
```

```python
players.head()
```

```python
# Deleting uneccesary columns
del players["Unnamed: 0"]
del players["Rk"]
```

```python
players["Player"].head(50)
```

```python
# Removing the astrick(*) beside players names
players["Player"] = players["Player"].str.replace("*", "", regex=False)
```

```python
players.head(50)
```

```python
players.head(20)
```

Above we can see that "Greg Anderson" is repeated four times. That is because he has played for all those teams in that year. So, we have to clean that up. Out of the four times "Greg Anderson" is repeated 3 enteries are for the 3 different teams he played and 1 is the total. We only want the total.

```python
players.groupby(["Player", "Year"])
```

```python
# function to drop the duplicates and assign the latest team
def single_row(df):
    if df.shape[0] == 1:
        return df
    else:
        row = df[df["Tm"] == "TOT"]
        row["Tm"] = df.iloc[-1,:]["Tm"]
        return row
    
players = players.groupby(["Player", "Year"]).apply(single_row)
```

```python
players.head(20)
```

```python
# dropping the index by one level
players.index = players.index.droplevel()
```

```python
# dropping the index by one level
players.index = players.index.droplevel()
```

```python
players[players["Player"] == "Greg Anderson"]
```

# Combining mvps and players data

```python
combined = players.merge(mvps, how="outer", on=["Player", "Year"])
# merging on outer because all the players data is not present in the mvps data.
```

```python
combined.head()
```

```python
# Checking if the merge worked properly
combined[combined["Pts Won"] > 0]
```

```python
# As there are NaN values only for the players who were never in the mvp voting we can replace those with 0s
combined[["Pts Won", "Pts Max", "Share"]] = combined[["Pts Won", "Pts Max", "Share"]].fillna(0)
```

```python
combined
```

# Combining teams and the above combined data

```python
# Reading the teams data
teams = pd.read_csv("teams.csv")
```

```python
teams
```

```python
teams.head(30)
```

We can see that some rows have division names in the table. That happened when scraping the teams data. So, that has to be fixed.

```python
# Removing division names from the data
teams = teams[~teams["W"].str.contains("Division")]
```

```python
# Removing the astrick(*) from the team names
teams["Team"] = teams["Team"].str.replace("*", "", regex=False)
```

```python
# Splitting into two new columns to get rid of "()" after a team name
teams[["team1", "tmp"]] = teams["Team"].str.split("(", expand=True)
```

```python
# Deleting the Team and tmp to assign the new team name to Team later
del teams["Team"]
del teams["tmp"]
```

```python
# Removing the space left at the end of the team names
teams["team1"] = teams["team1"].str.rstrip()
```

```python
# Assigning the cleaned team names to Team
teams["Team"] = teams["team1"]
```

```python
# Deleting the temporary team names
del teams["team1"]
```

```python
teams.head()
```

```python
teams["Team"].unique()
```

```python
combined["Tm"].unique()
```

As we can see the team name is full team name in one dataframe and only an abbrevation in the other. So, we have to fix this. To fix it we need to get new data where full team names are present in one column and the team abbrevation is present in another column.

```python
# Mapping the abbreviation and the name in a dictionary
nicknames = {}

with open("nicknames.csv") as f:
    lines = f.readlines()
    for line in lines[1:]:
        abbrev, name = line.replace("\n", "").split(",")
        nicknames[abbrev] = name
```

```python
nicknames
```

```python
combined["Team"] = combined["Tm"].map(nicknames)
```

```python
combined.head()
```

```python
# Merging the combined and teams data
stats = combined.merge(teams, how="outer", on=["Team", "Year"])
```

```python
stats
```

# Cleaning up the final dataset

```python
del stats["Unnamed: 0"]
```

```python
stats.dtypes
```

Many numerical columns are stored as strings by pandas. So, we have to convert them to numerical columns

```python
# Converting to numerical columns
stats = stats.apply(pd.to_numeric, errors="ignore")
```

```python
stats.dtypes
```

```python
# GB (Games Back) should also be a numeric column so checking what's wrong with it
stats["GB"].unique()
```

```python
# Because of the "-" the column is not numerical. So, replacing it with 0.
stats["GB"] = stats["GB"].str.replace("—", "0")
# The keyboard "-" didn't work so, had to copy paste form the output above then it got replaced.
```

```python
stats["GB"].unique()
```

```python
# Converting GB also to a numerical column
stats["GB"] = pd.to_numeric(stats["GB"])
```

```python
stats.dtypes
```

```python
# Storing the final cleaned data in a csv file
stats.to_csv("player_mvp_stats.csv")
```
